```python
# Pandas - Data manipulation and analysis library
import pandas as pd
# NumPy - mathematical functions on multi-dimensional arrays and matrices
import numpy as np

import seaborn as sns
# Matplotlib - plotting library to create graphs and charts
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
# discharging_rate_lag3
df_train['discharging_rate_lag3'].hist(bins = 20)
plt.xlabel('discharging_rate_lag3')
plt.ylabel('Frequency')
plt.xlim([-10, 10])
plt.show()
df_train['discharging_rate_lag3'].describe()
```

```python
# charge_cycle_time_below_12
df_train['charge_cycle_time_below_12'].value_counts().plot(kind='bar')
plt.show()
df_train['charge_cycle_time_below_12'].value_counts()
```

```python
# charging_rate_lag3
df_train['charging_rate_lag3'].hist(bins = 20)
plt.xlabel('charging_rate_lag3')
plt.ylabel('Frequency')
plt.xlim([-10, 10])
plt.show
df_train['charging_rate_lag3'].describe()
```

```python
# avg_volt_change_charging
df_train['avg_volt_change_charging'].hist(bins = 20)
plt.xlabel('avg_volt_change_charging')
plt.ylabel('Frequency')
plt.show()
df_train['avg_volt_change_charging'].describe()
```
